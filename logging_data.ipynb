### Eiso errors are symmetric, so if we use only the lower error and ask for lower and upper errors, you should get consistent answers with had you used both the lower and upper errors.

````python
param = 'eiso'
out1 = to_log(  x = df[param], 
               xerr=[df[param+'_err_low'], 
                     df[param+'_err_up']], 
               which='both', 
               errTypeReturn='ci'
            )

out1
````

````python
param = 'eiso'
out2 = to_log(  x = df[param], 
               xerr = df[param+'_err_low'], 
               which='both', 
               errTypeReturn='ci'
            )

out2
````

````python
out1[1] == out2[1], out1[2] == out2[2]
````

### Our epeakRest errors are NOT symmetric. So if we only pass the lower error, we do not expect to get the same as had we passed both the lower and upper errors. 

````python
param = 'epeakRest'
out1 = to_log(  x = df[param], 
               xerr=[df[param+'_err_low'], 
                     df[param+'_err_up']], 
               which='both', 
               errTypeReturn='ci'
            )

out1
````

````python
param = 'epeakRest'
out2 = to_log(  x = df[param], 
               xerr=df[param+'_err_low'], 
               which='both', 
               errTypeReturn='ci'
            )

out2
````

````python
out1[1] == out2[1], out1[2] == out2[2]
````
